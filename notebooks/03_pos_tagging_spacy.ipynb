```python
import os
os.chdir("..")
from src.pos_tagging.pos_tagging_df import pos_tagging_df
import pandas as pd
#%pip install spacy
#!python -m spacy download es_core_news_md

df = pd.read_csv('data/processed/resenas_clean.csv', sep=';')
df = df[['business_name', 'review_rating', 'review_text']]
df.sample(2)
```

```python
todos_los_registros = []

for lugar, grupo in df.groupby("business_name"):
    textos = grupo["review_text"].tolist()
    resultados_lugar = pos_tagging_df(lugar, textos)
    todos_los_registros.extend(resultados_lugar)

df_resultados = pd.DataFrame(todos_los_registros)
df_resultados =  df_resultados[df_resultados['categoria'] != 'MISC']
```
